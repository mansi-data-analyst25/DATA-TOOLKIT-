# QUES 1: What is NumPy, and why is it widely used in Python?

ANS: NumPy (short for Numerical Python) is a powerful open-source library in Python used for numerical computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays efficiently.

#Key Features of NumPy:

1. N-dimensional array object (ndarray):

Much more efficient than Python's built-in lists for numerical data.

Supports operations on entire arrays without explicit loops (vectorization).

2. Broadcasting:

Allows arithmetic operations on arrays of different shapes in a flexible way.

3. Mathematical functions:

Provides a wide range of mathematical operations like linear algebra, statistics, Fourier transforms, etc.

4. Integration with C/C++/Fortran:

Useful for performance-critical computations.

5. Tools for working with memory-mapped files:

Efficient handling of large datasets that do not fit into memory.

#Why is NumPy widely used?

> Performance:

Operations on NumPy arrays are implemented in C, making them much faster than equivalent Python code using loops.

> Convenience:

Provides simple syntax and powerful operations that are hard to implement efficiently with plain Python lists.

> Foundation for scientific computing:

Many other Python libraries (like SciPy, pandas, scikit-learn, TensorFlow, PyTorch) are built on top of or depend on NumPy.

> Interoperability:

Can easily integrate with other tools and libraries.

> Community support:

Well-documented, mature, and widely used in academia and industry.




```python
import numpy as np

# Create a 1D array
a = np.array([1, 2, 3, 4, 5])

# Perform element-wise operations
b = a * 2
print(b)

# Create a 2D array (matrix)
matrix = np.array([[1, 2], [3, 4]])
print(np.linalg.inv(matrix))  # Inverse of the matrix
```

# QUES 2: How does broadcasting work in NumPy?

ANS: Broadcasting is one of the most powerful concepts in NumPy — it allows arrays of different shapes to be used in arithmetic operations without the need to explicitly reshape or replicate data.

#Why is broadcasting useful?

> It avoids unnecessary copying of data (saves memory).

> It enables vectorized operations (makes code faster and simpler).

> It eliminates the need for explicit looping in many cases.

#How does it work?
When performing operations on two arrays, NumPy compares their shapes element-wise:

> If the dimensions match, or one of them is 1, they are compatible.
> If the dimensions are incompatible, NumPy will raise an error.

If compatible:

NumPy "broadcasts" the smaller array across the larger one so they behave as if they had the same shape.

#Broadcasting rules (simplified):

> If arrays have different ranks (number of dimensions), the shape of the smaller array is left-padded with ones.

> Then, NumPy compares dimensions from last to first:

If dimensions are equal → fine.

If one dimension is 1 → stretch it to match the other.

If they are unequal and neither is 1 → error.

#Examples:

1. Adding a scalar to an array
2. Adding arrays with compatible shapes
3. When broadcasting fails

```python
# Adding a scalar to an array

import numpy as np

a = np.array([1, 2, 3])
b = 10

result = a + b
print(result)
```

```python
# Adding arrays with compatible shapes

a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

result = a + b
print(result)
```

```python
# When broadcasting fails

a = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20])

# This will raise an error:
result = a + b
```

# QUES 6: What are the differences between NumPy arrays and Python lists?

ANS: Here’s a clear comparison of NumPy arrays vs Python lists.

#key differences:

1. Data Type Consistency:

> NumPy Array - Elements must all be of the same data type (int, float, etc.).

> Python List - Can hold mixed data types (e.g., ["text", 10, 3.14]).

2. Performance:

> NumPy Array - Much faster for numerical computations due to optimized C-based implementation.

> Python List - Slower because each element is a generic Python object.

3. Functionality & Operations:

> NumPy Array - Supports vectorized operations (e.g., element-wise addition, multiplication).

> Python List - Does not support vectorized math; requires manual iteration.

4. Memory Usage:

> NumPy Array - More memory efficient due to compact storage of homogeneous data.

> Python List - Consumes more memory since it stores references to objects.

5. Multidimensional Support:

> NumPy Array - Supports multidimensional arrays easily (matrices, tensors).

> Python List - Can simulate multi-dimensions (e.g., list of lists), but it's less efficient and more error-prone.

6. Built-in Functions:

> NumPy Array - Comes with a wide range of powerful mathematical, statistical, and linear algebra functions.

> Python List - Requires manual implementation or external libraries for most numerical operations.





```python
# Functionality & Operations: NumPy Array

import numpy as np
a = np.array([1, 2, 3])
print(a * 2)
```

```python
# Functionality & Operations: Python List

a = [1, 2, 3]
print([x * 2 for x in a])
```

# QUES 8: What does the term “vectorized operation” mean in NumPy?

ANS: A vectorized operation in NumPy refers to performing operations on entire arrays (vectors, matrices) at once, without using explicit loops in Python.

Instead of processing one element at a time (looping), NumPy performs the operation on the whole array simultaneously, using highly optimized C code under the hood.

This leads to faster and more concise code.

#Why Vectorization Matters?

> Faster — avoids slow Python loops.

> Cleaner code — more readable, less boilerplate.

> Memory efficient — uses optimized low-level operations.

#Examples of Vectorized Operations in NumPy:

| Operation                   | Example                   |
| --------------------------- | ------------------------- |
| Element-wise addition       | `c = a + b`               |
| Element-wise multiplication | `c = a * b`               |
| Element-wise division       | `c = a / b`               |
| Mathematical functions      | `np.sqrt(a)`, `np.sin(a)` |
| Logical operations          | `a > 0`, `a == b`         |



```python
# Without vectorization (Python list + loop)

a = [1, 2, 3, 4]
b = [x * 2 for x in a]  # Using a loop (list comprehension)
print(b)
```

```python
# With vectorization (NumPy array)

import numpy as np
a = np.array([1, 2, 3, 4])
b = a * 2  # Vectorized operation (no explicit loop)
print(b)
```

# QUES 13: Why is handling missing data important in Pandas?

ANS: Handling missing data is crucial in Pandas (and in data analysis in general) because missing values can significantly affect the quality, accuracy, and interpretability of your results.

#Why it matters:

1.  Ensures Accurate Calculations:

> Many analytical functions (mean, sum, correlation, machine learning models) will produce incorrect or misleading results if missing data is not handled properly.

> Example: The mean of a column with NaN values can be distorted.

2. Prevents Errors in Processing:

> Some functions or algorithms cannot handle missing values and will throw errors.

> Many ML models require complete datasets — they won’t fit if NaNs are present.

3. Maintains Data Integrity:

> Ignoring missing data can lead to biased conclusions.

> Example: If income data is missing mostly for high earners, simply dropping rows may under-represent that group.

4. Helps Identify Data Quality Issues:

> Missing data may reveal data collection problems, data entry errors, or incomplete sources.

> Analyzing why data is missing can give important insights (missing at random, missing not at random, etc.).

5. Required for Proper Visualization:

Many visualization tools and plots (heatmaps, scatterplots, pairplots) will either fail or misrepresent the data if missing values are not addressed.

#Typical Techniques to Handle Missing Data in Pandas:

| Method                                           | Function                  | When to Use                                                |
| ------------------------------------------------ | ------------------------- | ---------------------------------------------------------- |
| Drop rows/columns with missing data              | `dropna()`                | When missing data is minimal and non-critical              |
| Fill with constant (e.g., 0, 'Unknown')          | `fillna()`                | For categorical/text data or known default values          |
| Fill with summary statistic (mean, median, mode) | `fillna(df.mean())`, etc. | For numeric data when preserving distribution is important |
| Interpolation                                    | `interpolate()`           | For time-series or continuous data                         |
| Flag missing data                                | Create indicator column   | When you want models to learn that data was missing        |

# QUES 17: Explain the difference between apply() and map() in Pandas.

ANS: apply() and map() are both very useful in Pandas, but they serve slightly different purposes and are used in different contexts.

#Clear breakdown:

#map()

> Works only on Series.

> Applies a function element-wise to each value in the Series.

> You can pass:

A function

A dictionary (for value mapping)

A Series

#apply()

> Works on both Series and DataFrames.

> More general-purpose than map().

> On Series → behaves similarly to map() (element-wise).

> On DataFrame → applies a function along an axis (rows or columns).

| Feature             | `map()`                                 | `apply()`                                                                |
| ------------------- | --------------------------------------- | ------------------------------------------------------------------------ |
| Used with           | **Series** only                         | **Series** and **DataFrame**                                             |
| Purpose             | Element-wise transformation of a Series | Apply a function along an axis (rows/columns) or to elements in a Series |
| Function applied to | Each individual element of the Series   | Series elements (if used on Series), rows/columns (if used on DataFrame) |
| Common use case     | Value mapping, element-wise operations  | Row-wise/column-wise operations, complex functions                       |




```python
import pandas as pd

s = pd.Series([1, 2, 3, 4, 5])

# Apply a function to each element
s.map(lambda x: x ** 2)
```

```python
# Mapping values with a dictionary

s = pd.Series(['cat', 'dog', 'rabbit'])
s.map({'cat': 'kitten', 'dog': 'puppy'})
```

```python
# Example on Series

s = pd.Series([1, 2, 3, 4, 5])
s.apply(lambda x: x ** 2)
```

```python
# Example on DataFrame

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Apply function to each column (axis=0 is default)
df.apply(sum)

# Apply function to each row
df.apply(lambda row: row['A'] + row['B'], axis=1)
```

# QUES 21: Why is NumPy’s array slicing faster than Python’s list slicing?

ANS: Excellent question — NumPy array slicing is significantly faster than Python list slicing, and this comes down to differences in how data is stored and accessed in memory.

#Here’s why:

1. Memory Layout:

> NumPy arrays are stored as a single, contiguous block of memory (just like C arrays).

> Python lists are collections of references to objects scattered in memory (each element is a pointer to an object).

Because NumPy arrays are tightly packed, slicing does not require copying or dereferencing each element — it just creates a view of the original memory block.

2. Views vs Copies:

> NumPy slicing returns a view (no data copied), unless explicitly asked to copy.

> Python list slicing creates a new list (copies data).

3. Underlying Implementation:

| Feature               | NumPy array                 | Python list          |
| --------------------- | --------------------------- | -------------------- |
| Memory layout         | Contiguous block            | Array of pointers    |
| Slice operation       | View (pointer manipulation) | New list with copies |
| Speed                 | Very fast                   | Slower, linear copy  |
| Optimized for numeric | Yes                         | No                   |

4. No Type Checking:

> NumPy arrays are homogeneous → all elements are the same type.

This allows optimized, type-specific code.

> Python lists are heterogeneous → each element could be anything → more overhead when slicing.

```python
# Views vs Copies

import numpy as np

a = np.arange(1_000_000)
b = a[1000:2000]  # This is just a view → fast!

# vs. Python list
lst = list(range(1_000_000))
sub_lst = lst[1000:2000]  # Data is copied → slower
```

```python
lst = list(range(1_000_000))

# Sublist → creates a copy!
sub_lst = lst[1000:2000]

# Modify sublist — this does NOT affect lst
sub_lst = [x * 2 for x in sub_lst]

# lst remains unchanged
print(lst[995:1005])
```
